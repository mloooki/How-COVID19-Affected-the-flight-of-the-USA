```python
data_df.loc[data_df.CANCELLATION_CODE == "A", "CANCELLATION_CODE"] = "carrier"
data_df.loc[data_df.CANCELLATION_CODE == "B", "CANCELLATION_CODE"] = "weather"
data_df.loc[data_df.CANCELLATION_CODE == "C", "CANCELLATION_CODE"] = "NAS"
data_df.loc[data_df.CANCELLATION_CODE == "D", "CANCELLATION_CODE"] = "security"
```

```python
# Convert FL_DATE to datatime
data_df['FL_DATE'] = pd.to_datetime(data_df['FL_DATE'])
```

```python
# Convert CANCELLATION_CODE to category
data_df['CANCELLATION_CODE'] = data_df['CANCELLATION_CODE'].astype('category')
```

```python
# Convert YEAR, MONTH, DAY_OF_MONTH, DAY_OF_WEEK to string
data_df['YEAR'] =data_df['YEAR'].astype('str')
data_df['MONTH'] =data_df['MONTH'].astype('str')
data_df['DAY_OF_MONTH'] =data_df['DAY_OF_MONTH'].astype('str')
data_df['DAY_OF_WEEK'] =data_df['DAY_OF_WEEK'].astype('str')
```

```python
# get the day of the flight from FL_DATE
data_df['day'] = data_df.FL_DATE.dt.strftime('%A')
```

### What is the structure of your dataset?

The dataset contains 862,644 fligths with 19 variable. The dataset structure is the following:
#### 1- Airport 
<li> ORIGIN = Origin Airport </li>
<li> DEST = Destination Airport</li>

#### 2- Trip info
<li>FL_DATE = Fly Date </li>
<li>DAY_OF_WEEK = Number of day of the week (where 1 = Monday and 7 = Sunday) </li>
<li>DAY_OF_MONTH = number of day of the munth </li>
<li>AIR_TIME = Flight Time, in Minutes </li>
<li> DISTANCE = Distance between airports (miles)</li>
and I added 'day' to detemine the day (Sturday, Sunday ..ect)

#### 3- Delay 
<li>DEP_DELAY = Difference in minutes between scheduled and actual departure time. Early departures show negative numbers.</li>
<li>ARR_DELAY = Difference in minutes between scheduled and actual arrival time. Early arrivals show negative numbers. </li>
<li>CARRIER_DELAY = Carrier Delay, in Minutes </li>
<li>WEATHER_DELAY = Weather Delay, in Minutes </li>
<li>NAS_DELAY = National Air System Delay, in Minutes </li>
<li>SECURITY_DELAY = Security Delay, in Minutes </li>
<li>LATE_AIRCRAFT_DELAY = Late Aircraft Delay, in Minutes </li>

#### 4- Cancellations and Diversions
<li>CANCELLED = Cancelled Flight Indicator (1=Yes)</li>                
<li>CANCELLATION_CODE = Reason for Cancellation (A = carrier, B = weather, C = NAS, D = security) </li>       
<li>DIVERTED = Diverted Flight Indicator (1=Yes)</li>

### What is/are the main feature(s) of interest in your dataset?

The main feature of the dataset are:

## Univariate Exploration



```python
plt.figure(figsize = [12, 5])
plt.hist(data= data_df , x = 'DEP_DELAY');
plt.ylabel('count');
plt.xlabel('Departure delay');
```

```python
# zoom in to privent outlirs
plt.figure(figsize = [12, 5])
plt.hist(data= data_df , x = 'DEP_DELAY',bins=10);
plt.xlim((-100,370));
plt.ylabel('count');
plt.xlabel('Departure delay');
```

We can see from the previos figure there is outliers of Departure delay. and i will remove any trip with Departure delay over 360 min.

```python
outliers= data_df[data_df.DEP_DELAY > 360] # save outliers before delete them.
```

```python
data_df.drop(data_df[data_df.DEP_DELAY > 360].index, inplace=True)
```

```python
data_df.to_csv('Flights_Reporting_Cleaned.csv',index=False) # save the cleaned dataset
outliers.to_csv('outliers.csv',index=False) # save outliers for future work.
```

```python
plt.figure(figsize = [12, 5])
bins = np.arange(-100,data_df['ARR_DELAY'].max()+40,40)
plt.hist(data= data_df , x = 'ARR_DELAY',bins=bins);
plt.xlim((-100,370));
plt.ylabel('count');
plt.xlabel('Arrival delay');
```

Booth Arrival delay and Departure delay almost on the same interval (from -50 to 300). but we can see some values arrival delay start from -100 and up to about 330.
